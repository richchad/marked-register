## Marked register digitisation 
 

### Brief
For every UK  election (local or national) polling stations are provided with registers of the eligble voters for the respective polling station. As electors vote, their names are 'marked' to indicate such, ensuring they are not able to cast a vote twice. 

Marked registers are retained by local authorities and can be provided to political parties for a fee. Political parties purchase photocopies of the marked registers and members volunteer their time for data entry, a single ward (a constituency may have 8-12 wards) can take up to 10 hours to process. 

The high time investment associated with digitising marked registers results in only a portion of the data being processed, with a priority on digitising parts of the data that are expected to be more valuable (for example in a marginal constituency, where marked register data can have a genuine impact on campaign and targeting strategies).

<b>This project seeks to automate the process of digitising marked registers.</b>

Automating the process of digitising marked registers can save thousands of hours of volunteer labour. It will also allow for the creation of a nationwide, digitised marked register, which would represent the most comprehensive database of election turnout. 

That this dataset would be valuable is obvious. Two valuable use cases are:
1. Detailed analysis of voter turnout and what drives it, insights into how to promote civic engagement.
2. Identifying voter fraud.


#### Data protection
All marked registers are protected by GDPR - the purpose of this readme is to demonstrate my work so far without revealing any sensitive information. All additional work is stored in private repository.

# Summary of digitisation strategy

The plan is to crop each voter from the marked register, keeping record of the order they appear in the register. Given a csv of the electoral register, use the images order and pytesseracts interpretation of the image text to map images to the correct elector in the csv. Apply machine learning techniques to labelled (voted/ not voted) images to create a model that predicts whether the register has been marked (a vote has been cast).

### Acquire data
1. Convert each page of scanned mark register to png image.
2. Deskew images (using correct_skew_folder.py)
3. Run pytesseract on each page to identify every symbol and the coorindates of their bounding boxes.
4. Determine x coodrinate to split page (using DBScan, see below).
5. Cluster symbols by line (DBScan)
6. For each page half, use median left position of all clusters and median bottom position of each cluster to identify bottom left corner of crop (crop width and height all the same).
7. Save cropped images to disk (name image by order).
9. Assign labels by allocating to subfolders (voted, not voted, absentee voter, redundant image etc)
10. Refine crop dimensions (after running preliminary models)

### Map image to elector (notes)
1. Pytesseract is not always accurate, it performs better with a distinct contrast between black and whites. It may be necessary to heighten the images contrast to improve performance (a consistent contrast between images will also help modelling).
2. Regex
3. Translate misclassified symbols based on order (should always be polling number first, B=8 for example).
4. Classify in tranches, map 'easy classifications' with text then use ordering to establish mapping for images more difficult to map.


### Modelling on images (notes)
1. Labels assigned manually, will need to assign a lot of labels, creating interface for one click labelling would be helpful.
2. Crop coordinates and dimensions will need to be optimised. Images may need to be manipulated or recropped to optimise for machine learning.
3. Given the aim is to identify the presence of a horizontal(ish) line, there may be other numeric representations of the image (other than straight array to dataframe conversion) that will result in higher model accuracy.
3. Pytesseract can interpret lines as dashes, this interpretation could be used to establish whether vote has been cast.
4. Probability threshold for classification could be used to classify a majority of voters with edge cases being presented to humans for classification.
3. Model, model, model...

### Deskew images
Deskew script calculates approximate bounding box of black pixels and rotates to be straight, as a result it does not work with pages that have a dark edge (block black edges show up when photocopied page has not been lined up properly). Workaround options:

- DBSCAN on black pixels, whiten them if they pass a threshold of density (associated with block of black, but not text).
- Plot horizontal lines of varying angles, use lines that intersects least amount of black pixels to find appropriate deskew angle

### Get symbol coordinates for each page


```python
# Get bounding box estimates of every letter in image
def letter_df(image):
    "Input file path of jpeg image"
    #apply pytesseract to png file
    boxes=pytesseract.image_to_boxes(image)
    
    #parse information into data frame
    boxes=boxes.split('\n')
    for i in range(len(boxes)):
        boxes[i]=boxes[i].split(' ')
    data_dict={'symbol':[i[0] for i in boxes],
               'left':[int(i[1]) for i in boxes],
               'bottom':[int(i[2]) for i in boxes] ,
               'right': [int(i[3]) for i in boxes],
               'top': [int(i[4]) for i in boxes]}
    boxes = pd.DataFrame(data_dict)
    
    #add dummy column for one variable clustering
    boxes['dummy']=[0]*len(boxes)
    
    #identify all digits in dataframe
    boxes['is_digit']=[(i.isdigit())*1 for i in boxes['symbol']]
    
    #identify letter or number
    #bounding boxes of punctuation are not aligned with characters and digits
    #isolating characters and numbers makes identifying lines easier
    boxes['is_character']=[(i.isdigit() | i.isalpha())*1 for i in boxes['symbol']]
    
    return boxes
```

## Split page in centre on x axis
This process takes into account that the centre point will not be consistent between pages.

```python
#subset table for digits only
page_numbers=page[page.is_digit==True]
pnc=page_numbers[(page_numbers.left>400) & (page_numbers.left<1300)]  #page_numbers_cropped
```

```python
#run dbscan on digits x axis position
dbscan = DBSCAN(eps=25, min_samples=15)
dbscan.fit(pnc[['left','dummy']])

pnc['labels']=dbscan.labels_

split=(pnc[pnc['labels']==0]['right'].max()+pnc[pnc['labels']==1]['left'].min())/2

page['is_left_side'] = [(i<=split)*1 for i in page['right']]
```

```python
#step by step process illustrated with plots
fig, ax = plt.subplots(nrows=2, ncols=2 ,figsize=(12, 20), sharey=True, sharex=True)

fig.suptitle('Split page with dbscan', fontsize=40)

sns.scatterplot(x=page.left, y=page.bottom , hue=page.is_digit, ax=ax[0,0])
ax[0,0].set_title('All symbols')

sns.scatterplot(x=page_numbers['left'], y=page_numbers.bottom,ax=ax[0,1])
ax[0,1].set_title('All digits')

ax[1,0].set_title('Clustered digits (cast vote)')
sns.scatterplot(x=pnc['left'], y=pnc.bottom, hue=pnc['labels'], ax=ax[1,0])

ax[1,1].set_title('All symbols + intercept line')
sns.scatterplot(x=page.left, y=page.bottom)
plt.plot([split, split], [page_numbers.bottom.min(), page_numbers.bottom.max()], 'k-', color = 'r')


plt.show()
```

### Get coordinates for each line (on each page half)

```python
def get_line_info(letter_df):
    """Runs DBscan on y coordinates of characters to cluster by line, adds label to original data frame"""
    
    #Set all line labels to not classified
    letter_df['line_labels'] = -1
    
    #Perform on left side first
    info = letter_df[(letter_df['is_character']==1) & (letter_df['is_left_side']==1)]
    
    #run dbscan on y coordinates only
    dbscan = DBSCAN(eps=7, min_samples=8) #eps effective between 7 and 11(ish)
    dbscan.fit(info[['top','dummy']])
    
    #order matters - use top coordinate to relabel clusters in order top down
    #assign labels, subset for labelled only
    info['line_labels']=dbscan.labels_
    info=info[info['line_labels']>=0]
    
    #sort values, append cluster label to list in 
    #order of heighest to lowest position on page
    info = info.sort_values(by='top', ascending=False)
    order=[]
    for i in info['line_labels']:
        if i not in order:
            order.append(i)
    
    true_order=range(len(order))
    paired = zip(order, true_order)
    pair_dict={i:j for i,j in paired}
    info['true_line_labels']=[pair_dict[i] for i in info['line_labels']]
    
    #assign ordered line labels to letter_df.line_labels
    for i in info.index:
        letter_df.loc[i ,'line_labels']= info.loc[i, 'true_line_labels']
    
    #starting line count for right side
    start= max(true_order)+1
    
    #repeat for right side
    info = letter_df[(letter_df['is_character']==1) & (letter_df['is_left_side']==0)]
    dbscan = DBSCAN(eps=7, min_samples=8) 
    dbscan.fit(info[['top','dummy']])
    info['line_labels']=dbscan.labels_
    info=info[info['line_labels']>=0]
    info = info.sort_values(by='top', ascending=False)
    order=[]
    for i in info['line_labels']:
        if i not in order:
            order.append(i)
    true_order=range(start, start+len(order))
    paired = zip(order, true_order)
    pair_dict={i:j for i,j in paired}
    info['true_line_labels']=[pair_dict[i] for i in info['line_labels']]
    for i in info.index:
        letter_df.loc[i ,'line_labels']= info.loc[i, 'true_line_labels']
    
    return letter_df
```

```python
page=get_line_info(page)
```

## Crop out each voter
For each half of page use median bottom coordinate of each cluster and median left coordinate of all clusters to determine the bottom left coordinate of each crop.

Each crop is the same size (necessary for predictive modelling).

```python
def cropped_images(page, image):
    
    #get letter_df and subset for identified lines
    info=page
    info=info[info['line_labels']>=0]
    
    #get image as array
    image_array = cv2.imread(image)
    
    #setup dictionary for storing images
    images={}
    
    #image height
    height=image_array.shape[0]
    
    #unique lines ordered
    left_info= round(info[info['is_left_side']==1].groupby(['line_labels']).min()['left'].median())-15
    right_info=round(info[info['is_left_side']==0].groupby(['line_labels']).min()['left'].median())-15
    left_lines=np.unique(info[info['is_left_side']==1]['line_labels'])
    right_lines=np.unique(info[info['is_left_side']==0]['line_labels'])
    
    #find left and bottom coordinate of each unique line
    for line in left_lines:
        line_info=info[(info.line_labels==line )]
        bottom=height-round(line_info.bottom.median())+5
        right= line_info.right.max()+5
        #top=height-line_info.top.max()-5 (will use fixed sizing instead)
        images['pic{}'.format(line)]= [image_array[bottom-40:bottom, left_info+80:left_info+600] ,line_info]
        
    for line in right_lines:
        line_info=info[(info.line_labels==line )]
        bottom=height-round(line_info.bottom.median())+5
        images['pic{}'.format(line)]= [image_array[bottom-40:bottom, right_info+80:right_info+600] ,line_info]
        
        #if statement drops some images that aren't voters 
        #600 threshold def scaleable, not sure about 785 one
        #will need to check to see if these scale between versions
        
        #if ((right-left>600) & (right-left<785)):
            
            #append image and coordinates of letters to dictionary
            #will need to check if 45 and 780 sizing scales
            #need all images same size for training
        #images['pic{}'.format(line)]= [image_array[bottom-40:bottom, left:left+780] ,line_info]
    
    return images
```

```python
def strings_by_line(page, image):
    
    df=page
    df=df[df.line_labels>=0].sort_values(by=['line_labels','left'])
    strings=[]

    for line in np.unique(df.line_labels):
        word=''
        for letter in df[(df.line_labels==line)]['symbol']:
            word= word +letter
        strings.append(word)
    return strings
```

```python
example_crop=cropped_images(page,'example/example_page10.png' )
image_as_strings= strings_by_line(page,'example/example_page10.png' )
```
